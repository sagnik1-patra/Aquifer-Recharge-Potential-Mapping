```python
import os
import json
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Aquifer Recharge Potential Mapping"
ARCHIVE_DIR = os.path.join(BASE_DIR, "archive")

DISTRICT_FILE = os.path.join(ARCHIVE_DIR, "district wise rainfall normal.csv")
HISTORICAL_FILE = os.path.join(ARCHIVE_DIR, "rainfall in india 1901-2015.csv")

os.makedirs(BASE_DIR, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

district_df = pd.read_csv(DISTRICT_FILE)
historical_df = pd.read_csv(HISTORICAL_FILE)

district_df.columns = district_df.columns.str.lower().str.strip()
historical_df.columns = historical_df.columns.str.lower().str.strip()

# ============================================================
# FEATURE ENGINEERING
# ============================================================

district_rain_cols = district_df.select_dtypes(include=np.number).columns
district_df["mean_rainfall"] = district_df[district_rain_cols].mean(axis=1)
district_df["std_rainfall"] = district_df[district_rain_cols].std(axis=1)

district_df["recharge_index"] = (
    district_df["mean_rainfall"] / district_df["mean_rainfall"].max()
)

features = ["mean_rainfall", "std_rainfall"]
target = "recharge_index"

X = district_df[features]
y = district_df[target]

# ============================================================
# SCALING & SPLIT
# ============================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ============================================================
# MODEL TRAINING
# ============================================================

model = RandomForestRegressor(
    n_estimators=300,
    max_depth=12,
    random_state=42
)

model.fit(X_train, y_train)

train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

train_r2 = r2_score(y_train, train_pred)
test_r2 = r2_score(y_test, test_pred)

# ============================================================
# 1 ACCURACY GRAPH
# ============================================================

plt.figure()
plt.plot(["Train", "Test"], [train_r2, test_r2], marker="o")
plt.title("Model Accuracy (R Score)")
plt.ylabel("R Score")
plt.grid(True)
plt.savefig(os.path.join(BASE_DIR, "accuracy_curve.png"))
plt.show()

# ============================================================
# 2 PREDICTION vs ACTUAL GRAPH
# ============================================================

plt.figure()
plt.scatter(y_test, test_pred)
plt.xlabel("Actual Recharge Index")
plt.ylabel("Predicted Recharge Index")
plt.title("Prediction vs Actual Comparison")
plt.grid(True)
plt.savefig(os.path.join(BASE_DIR, "prediction_vs_actual.png"))
plt.show()

# ============================================================
# 3 CORRELATION HEATMAP
# ============================================================

corr = district_df[features + [target]].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Feature Correlation Heatmap")
plt.savefig(os.path.join(BASE_DIR, "correlation_heatmap.png"))
plt.show()

# ============================================================
# 4 PREDICTION DISTRIBUTION GRAPH
# ============================================================

plt.figure()
plt.hist(test_pred, bins=20)
plt.title("Recharge Prediction Distribution")
plt.xlabel("Recharge Index")
plt.ylabel("Frequency")
plt.grid(True)
plt.savefig(os.path.join(BASE_DIR, "prediction_distribution.png"))
plt.show()

# ============================================================
# 5 SAVE RESULT CSV
# ============================================================

results_df = district_df.copy()
results_df["predicted_recharge"] = model.predict(X_scaled)

results_csv_path = os.path.join(BASE_DIR, "recharge_results.csv")
results_df.to_csv(results_csv_path, index=False)

# ============================================================
# 6 SAVE PREDICTION JSON
# ============================================================

predictions_json = {
    "model": "RandomForestRegressor",
    "train_r2": train_r2,
    "test_r2": test_r2,
    "predictions": results_df["predicted_recharge"].tolist()
}

json_path = os.path.join(BASE_DIR, "recharge_predictions.json")
with open(json_path, "w") as f:
    json.dump(predictions_json, f, indent=4)

# ============================================================
# 7 SAVE MODEL & SCALER
# ============================================================

with open(os.path.join(BASE_DIR, "recharge_model.pkl"), "wb") as f:
    pickle.dump(model, f)

with open(os.path.join(BASE_DIR, "recharge_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

# ============================================================
# DONE
# ============================================================

print(" ALL RESULTS GENERATED & SAVED")
print(" Output Directory:", BASE_DIR)
print(" Graphs + CSV + JSON + PKL created successfully")
```
